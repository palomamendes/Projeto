```python
if response_pais.status_code == 200 and response_ligas.status_code == 200 and response_artilheiros.status_code == 200:
    data_pais = response_pais.json()
    data_ligas = response_ligas.json()
    data_artilheiros = response_artilheiros.json()
else:
    print("Erro ao acessar a API")
```

```python
# Utilizando a função json_normalize para apresentar os resultados em forme de tabela
# E o metodo fillna para tratativa de valores ausentes

pais_full = pd.DataFrame(data_pais)
pais_full = pd.json_normalize(pais_full.result).fillna(0)
pais_full
```

```python
# Utilizando a função json_normalize para apresentar os resultados em forme de tabela
# E o metodo fillna para tratativa de valores ausentes

ligas_full = pd.DataFrame(data_ligas)
ligas_full = pd.json_normalize(ligas_full.result).fillna(0)
ligas_full
```

```python
# Utilizando a função json_normalize para apresentar os resultados em forme de tabela
# E o metodo fillna para tratativa de valores ausentes

artilheiros_full = pd.DataFrame(data_artilheiros)
artilheiros_full = pd.json_normalize(artilheiros_full.result).fillna(0) 
artilheiros_full
```

```python
# Renomeando o titulo das colunas
pais_full = pais_full.rename(columns={
    'country_key': 'id_pais',
    'country_name': 'nome_pais',
    'country_iso2': 'sigla_pais',
    'country_logo': 'logo_pais'
})
ligas_full = ligas_full.rename(columns={
    'league_key':'id_liga',
    'league_name':'nome_liga',
    'country_key': 'id_pais',
    'country_name': 'nome_pais',
    'league_logo': 'logo_liga',
    'country_logo': 'pais_logo'
})

artilheiros_full = artilheiros_full.rename(columns={
    'player_place':'colocacao',
    'player_name':'nome_jogador',
    'player_key': 'id_jogador',
    'team_name': 'time',
    'team_key': 'id_time',
    'goals': 'qtd_gols',
	'assists': 'qtd_assist',
	'penalty_goals': 'gols_penalti',
})

					
```

```python
# Tratando os tipos das variaveis
pais_full['id_pais'] = pais_full['id_pais'].astype('int')
pais_full['nome_pais'] = pais_full['nome_pais'].astype('string')
pais_full['sigla_pais'] = pais_full['sigla_pais'].astype('string')
pais_full['logo_pais'] = pais_full['logo_pais'].astype('string')
					
ligas_full['id_liga'] = ligas_full['id_liga'].astype('int')
ligas_full['nome_liga'] = ligas_full['nome_liga'].astype('string')
ligas_full['id_pais'] = ligas_full['id_pais'].astype('int')
ligas_full['nome_pais'] = ligas_full['nome_pais'].astype('string')
ligas_full['logo_liga'] = ligas_full['logo_liga'].astype('string')
ligas_full['pais_logo'] = ligas_full['pais_logo'].astype('string')

artilheiros_full['colocacao'] = artilheiros_full['colocacao'].astype('int')
artilheiros_full['nome_jogador'] = artilheiros_full['nome_jogador'].astype('string')
artilheiros_full['id_jogador'] = artilheiros_full['id_jogador'].astype('int')
artilheiros_full['time'] = artilheiros_full['time'].astype('string')
artilheiros_full['id_time'] = artilheiros_full['id_time'].astype('int')
artilheiros_full['qtd_gols'] = artilheiros_full['qtd_gols'].astype('int')
artilheiros_full['qtd_assist'] = artilheiros_full['qtd_assist'].astype('int')
artilheiros_full['gols_penalti'] = artilheiros_full['gols_penalti'].astype('int')

```
