```python
# Utilizando a função json_normalize para apresentar os resultados em forme de tabela
# E o metodo fillna para tratativa de valores ausentes
genero_full = pd.DataFrame(response_genero)
genero_full = pd.json_normalize(genero_full.genres).fillna(0)
genero_full
```

```python
# Utilizando a função json_normalize para apresentar os resultados em forme de tabela
# drop para excluir colunas que não serão utilizadas
# E o metodo fillna para tratativa de valores ausentes
filmes_full = pd.DataFrame(response_filmes)
filmes_full = pd.json_normalize(filmes_full.results).fillna(0)
filmes_full['genre_ids'] = filmes_full['genre_ids'].apply(lambda x: x[0] if x else None)

excluir_col_filmes = ['adult','backdrop_path','original_title','original_language','popularity','poster_path','release_date','video','vote_count']

filmes_full = filmes_full.drop(excluir_col_filmes, axis=1)


filmes_full
```

```python
# Utilizando a função json_normalize para apresentar os resultados em forme de tabela
# drop para excluir colunas que não serão utilizadas
# E o metodo fillna para tratativa de valores ausentes
series_full = pd.DataFrame(response_series)
series_full = pd.json_normalize(series_full.results).fillna(0)
series_full['genre_ids'] = series_full['genre_ids'].apply(lambda x: x[0] if x else None)
series_full['origin_country'] = series_full['origin_country'].apply(lambda x: x[0] if x else None)

excluir_col_series = ['adult','backdrop_path','original_name','original_language','popularity','poster_path','first_air_date','vote_count']

series_full = series_full.drop(excluir_col_series, axis=1)

series_full
```

```python
# Renomeando o titulo das colunas
genero_full = genero_full.rename(columns={
    'id': 'id_genero',
    'name': 'genero'
})
filmes_full = filmes_full.rename(columns={				
    'genre_ids': 'id_genero',
    'id': 'id_filme',
    'original_title': 'titulo_original',
    'overview': 'overview',
    'title':'titulo',
    'vote_average':'votacao_media'
})

series_full = series_full.rename(columns={
    'genre_ids': 'id_genero',
    'id': 'id_serie',
    'origin_country': 'pais_origem',
    'overview': 'overview',
    'name':'titulo',
    'vote_average':'votacao_media'
})

```

```python
# Relacionando as tabelas

join_generos = pd.merge(pd.merge(genero_full,filmes_full, on='id_genero', how='left'),series_full, on='id_genero', how='left') 


join_generos.rename(columns={'titulo_x': 'titulo'}, inplace=True)

join_generos['Categoria'] = join_generos.apply(lambda row: 'Filme' if pd.notna(row['id_filme']) else 'Serie', axis=1)

join_generos['titulo'] = join_generos['titulo'].fillna(join_generos['titulo_y'])

join_generos = join_generos.dropna(subset=['titulo','titulo_y'], how='all')

join_generos.drop(['id_genero','id_filme','overview_x','votacao_media_x','id_serie','pais_origem','overview_y','titulo_y','votacao_media_y'], axis=1, inplace=True)



join_generos
```

```python
# Tratando os tipos das variaveis
genero_full['id_genero'] = genero_full['id_genero'].astype('int')
genero_full['genero'] = genero_full['genero'].astype('string')
					
filmes_full['id_genero'] = filmes_full['id_genero'].astype('int')
filmes_full['id'] = filmes_full['id_filme'].astype('int')
filmes_full['overview'] = filmes_full['overview'].astype('string')
filmes_full['titulo'] = filmes_full['titulo'].astype('string')
filmes_full['votacao_media'] = filmes_full['votacao_media'].astype('float')

series_full['genre_ids'] = series_full['id_genero'].astype('int')
series_full['id'] = series_full['id_serie'].astype('int')
series_full['origin_country'] = series_full['pais_origem'].astype('string')
series_full['overview'] = series_full['overview'].astype('string')
series_full['titulo'] = series_full['titulo'].astype('string')
series_full['votacao_media'] = series_full['votacao_media'].astype('float')
```
